# テスト

## ヴィンテージ一NGワードを除外する

```python
# df_bulkに含まれているレコードでproduct_noise_judgeに含まれているノイズを含んでいるレコードが除外されているか ex. リーバイスビンテージクロージング
# df_bulkに含まれているレコードでproduct_noise_judgeに含まれているノイズを含んでいないレコードが除外されていないか ex. BIGE後期
```

```python
def test_1():
    assert len(
        df_bulk[
            ["リーバイスビンテージクロージング" in name for name in df_bulk["name"]]
        ]
    )
    assert not len(
        df_bulk_not_noise_ng_word[
            [
                "リーバイスビンテージクロージング" in name
                for name in df_bulk_not_noise_ng_word["name"]
            ]
        ]
    )


test_1()
```

```python
def test_2():
    assert len(df_bulk[["BIGE後期" in name for name in df_bulk["name"]]])
    assert len(
        df_bulk_not_noise_ng_word[
            ["BIGE後期" in name for name in df_bulk_not_noise_ng_word["name"]]
        ]
    )


test_2()
```

## ラインNGワードを除外する

```python
# df_bulk_not_noise_ng_wordに含まれているレコードでproduct_noise_judgeに含まれているノイズを含んでいるレコードが除外されているか ex. 501XX
# df_bulk_not_noise_ng_wordに含まれているレコードでproduct_noise_judgeに含まれているノイズを含んでいないレコードが除外されていないか ex. BIGE後期
```

```python
def test_3():
    assert len(
        df_bulk_not_noise_ng_word[
            ["501XX" in name for name in df_bulk_not_noise_ng_word["name"]]
        ]
    )
    assert not len(
        df_bulk_not_noise_ng_line[
            ["501XX" in name for name in df_bulk_not_noise_ng_line["name"]]
        ]
    )


test_3()
```

```python
def test_4():
    assert len(
        df_bulk_not_noise_ng_word[
            ["BIGE後期" in name for name in df_bulk_not_noise_ng_word["name"]]
        ]
    )
    assert len(
        df_bulk_not_noise_ng_line[
            ["BIGE後期" in name for name in df_bulk_not_noise_ng_line["name"]]
        ]
    )


test_4()
```
